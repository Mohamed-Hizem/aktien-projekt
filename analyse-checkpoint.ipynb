1-Loading Data with Python

a-Download data


```python
import yfinance as yf
import pandas as pd
import numpy as np
import os

# Aktien und Zeitraum definieren
TICKERS = ['AAPL', 'TSLA', 'MSFT', 'SAP.DE']
START    = '2020-01-01'
END      = '2024-12-31'

print(f'Lade Daten fur {len(TICKERS)} Aktien...')
raw = yf.download(TICKERS, start=START, end=END)

# Nur Schlusskurse extrahieren
close = raw['Close'].copy()
close.reset_index(inplace=True)
close.columns.name = None

print(f'Datenpunkte geladen: {len(close)}')
print(close.head())
```

b-Convert data to long format

```python
# Wide zu Long Format transformieren
df = close.melt(
    id_vars='Date',
    var_name='ticker',
    value_name='schluss'
)
df.rename(columns={'Date': 'datum'}, inplace=True)
df = df.dropna()          # fehlende Werte entfernen
df['datum'] = pd.to_datetime(df['datum'])
df['datum_str'] = df['datum'].dt.strftime('%Y-%m-%d')

# Vollstandige OHLCV-Daten laden (Open, High, Low, Close, Volume)
records = []
for ticker in TICKERS:
    t = yf.Ticker(ticker)
    hist = t.history(start=START, end=END)[['Open','High','Low','Close','Volume']]
    hist['ticker'] = ticker
    hist.reset_index(inplace=True)
    records.append(hist)

df_full = pd.concat(records, ignore_index=True)
df_full.rename(columns={
    'Date':'datum','Open':'open_kurs','High':'hoch',
    'Low':'tief','Close':'schluss','Volume':'volumen'
}, inplace=True)
os.makedirs("data", exist_ok=True)

# Speichern als CSV
df_full.to_csv('data/aktiendaten_roh.csv', index=False)
print('CSV gespeichert!')
print(df_full.head(8))
```

2-Set up an SQL database

a-Create and populate a database

```python
import sqlite3
os.makedirs("db", exist_ok=True)

# Verbindung herstellen (Datei wird automatisch erstellt)
conn = sqlite3.connect('db/aktien.db')
cursor = conn.cursor()

# Tabelle 'kurse' anlegen
cursor.execute('''
    CREATE TABLE IF NOT EXISTS kurse (
        id        INTEGER PRIMARY KEY AUTOINCREMENT,
        datum     TEXT    NOT NULL,
        ticker    TEXT    NOT NULL,
        open_kurs REAL,
        hoch      REAL,
        tief      REAL,
        schluss   REAL    NOT NULL,
        volumen   INTEGER,
        UNIQUE(datum, ticker)
    )
''')
# Datum in String umwandeln
df_full['datum'] = df_full['datum'].astype(str)
# Daten einfugen (bei Duplikaten uberschreiben)
df_full.to_sql('kurse', conn, if_exists='replace', index=False)

conn.commit()
print(f'Datensatze in DB: {cursor.execute("SELECT COUNT(*) FROM kurse").fetchone()[0]}')


```

b-SQL Queries for Analysis

```python
# Jahrliche Durchschnittskurse
query = """
SELECT
    ticker,
    strftime('%Y', datum) AS jahr,
    ROUND(AVG(schluss), 2) AS durchschnitt,
    ROUND(MIN(schluss), 2) AS minimum,
    ROUND(MAX(schluss), 2) AS maximum
FROM kurse
GROUP BY ticker, jahr
ORDER BY ticker, jahr;
"""

df = pd.read_sql_query(query, conn)
print(df.head())

```

```python
# Groeste Tagesgewinne pro Aktie
query = """
SELECT 
    ticker, 
    datum,
    ROUND(schluss, 2) AS kurs,
    ROUND((schluss - open_kurs) / open_kurs * 100, 2) AS tagesrendite_pct
FROM kurse
WHERE ((schluss - open_kurs) / open_kurs * 100) > 0
ORDER BY tagesrendite_pct DESC
LIMIT 10;
"""
df = pd.read_sql_query(query, conn)

print(df)
```

```python
# Handelsvolumen-Analyse

query = """
SELECT ticker,
    ROUND(AVG(volumen)) AS avg_volumen,
    MAX(volumen)        AS max_volumen,
    SUM(volumen)        AS gesamt_volumen
FROM kurse
GROUP BY ticker
ORDER BY avg_volumen DESC;
"""
df = pd.read_sql_query(query, conn)
print(df)
```

3-Analysis using Python

a-Calculate key figures

```python
df = df.sort_values(['ticker', 'datum'])

# Gleitende Durchschnitte (Moving Averages)
df['MA_7']  = df.groupby('ticker')['schluss'].transform(
    lambda x: x.rolling(window=7,  min_periods=1).mean())
df['MA_30'] = df.groupby('ticker')['schluss'].transform(
    lambda x: x.rolling(window=30, min_periods=1).mean())
df['MA_90'] = df.groupby('ticker')['schluss'].transform(
    lambda x: x.rolling(window=90, min_periods=1).mean())

# Tagliche Rendite in Prozent
df['rendite'] = df.groupby('ticker')['schluss'].pct_change() * 100

# Rollende 30-Tage Volatilitat (Standardabweichung der Renditen)
df['volatilitaet'] = df.groupby('ticker')['rendite'].transform(
    lambda x: x.rolling(window=30, min_periods=5).std())

df['datum'] = df['datum'].astype(str)
# In Datenbank speichern
conn = sqlite3.connect('db/aktien.db')
df[['datum','ticker','MA_7','MA_30','MA_90','rendite','volatilitaet']] \
  .to_sql('kennzahlen', conn, if_exists='replace', index=False)
conn.close()
print('Kennzahlen berechnet und gespeichert.')
```

b-Visualize price trends

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Aktienkurse 2020-2024 mit gleitenden Durchschnitten',
             fontsize=16, fontweight='bold', y=1.02)

farben = {'AAPL':'#1F4E79','TSLA':'#E74C3C','MSFT':'#27AE60','SAP.DE':'#8E44AD'}

for ax, ticker in zip(axes.flat, TICKERS):
    sub = df[df['ticker'] == ticker].copy()
    ax.plot(sub['datum'], sub['schluss'], color=farben[ticker],
            linewidth=1.2, label='Schlusskurs', alpha=0.8)
    ax.plot(sub['datum'], sub['MA_30'],
            color='orange', linewidth=2, linestyle='--', label='MA 30')
    ax.plot(sub['datum'], sub['MA_90'],
            color='red', linewidth=2, linestyle=':', label='MA 90')
    ax.set_title(ticker, fontweight='bold', fontsize=13)
    ax.set_xlabel('Datum')
    ax.set_ylabel('Kurs (USD / EUR)')
    ax.legend(fontsize=9)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.grid(True, alpha=0.3)


os.makedirs("charts", exist_ok=True)
plt.tight_layout()
plt.savefig('charts/kursverlauf_MA.png', dpi=150, bbox_inches='tight')
plt.show()
```

4-Simple forecast

a-Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import numpy as np

def prognose_linear(df, ticker_name, tage_zukunft=90):
    sub = df[df['ticker'] == ticker_name].dropna().copy()
    sub = sub.sort_values('datum')

    sub['datum'] = pd.to_datetime(sub['datum'])
    # Feature: Anzahl Tage seit Startdatum
    sub['tage'] = (sub['datum'] - sub['datum'].min()).dt.days

    X = sub[['tage']].values
    y = sub['schluss'].values

    # Modell trainieren
    modell = LinearRegression()
    modell.fit(X, y)

    # Modell evaluieren
    y_pred = modell.predict(X)
    mae = mean_absolute_error(y, y_pred)
    r2  = r2_score(y, y_pred)
    print(f'{ticker_name}: MAE={mae:.2f}  R2={r2:.4f}')

    # Zukunftsprognose
    letzter_tag = sub['tage'].max()
    zukunft_tage = np.array(
        [[letzter_tag + i] for i in range(1, tage_zukunft + 1)]
    )
    prognose = modell.predict(zukunft_tage)

    return sub, prognose, modell

df['datum'] = pd.to_datetime(df['datum'], utc=True).dt.tz_convert(None)
# Prognose fur Apple
sub_aapl, prog_aapl, modell_aapl = prognose_linear(df, 'AAPL', 90)
print(f'AAPL Prognose in 90 Tagen: ${prog_aapl[-1]:.2f}')
```

b-Prognose visualisieren

```python
def plot_prognose(sub, prognose, ticker_name):
    letztes_datum = sub['datum'].max()
    zukunft_daten = pd.date_range(
        start=letztes_datum + pd.Timedelta(days=1),
        periods=len(prognose), freq='B'   # 'B' = Werktage
    )

    fig, ax = plt.subplots(figsize=(13, 5))

    # Historische Kurse
    ax.plot(sub['datum'], sub['schluss'],
            color='#1F4E79', linewidth=1.2, label='Historisch')

    # MA-30 Linie
    ax.plot(sub['datum'], sub['MA_30'],
            color='orange', linewidth=1.5,
            linestyle='--', label='MA 30', alpha=0.7)

    # Prognose
    ax.plot(zukunft_daten, prognose,
            color='red', linewidth=2,
            linestyle='--', label='Prognose (90 Tage)')

    ax.axvline(letztes_datum, color='gray',
               linestyle=':', linewidth=1.5, label='Heute')

    ax.set_title(f'{ticker_name} - Kursverlauf & 90-Tage-Prognose',
                 fontweight='bold')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'charts/prognose_{ticker_name}.png', dpi=150)
    plt.show()

plot_prognose(sub_aapl, prog_aapl, 'AAPL')
```

Dashboard in Excel and Cognos 

```python
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.chart import LineChart, Reference
import pandas as pd, sqlite3

conn  = sqlite3.connect('db/aktien.db')
df    = pd.read_sql('SELECT * FROM kurse', conn)
df_kz = pd.read_sql('SELECT * FROM kennzahlen', conn)
conn.close()

# Zusammenfassung berechnen
summary = df.groupby('ticker').agg(
    start_kurs=('schluss', 'first'),
    end_kurs=('schluss', 'last'),
    max_kurs=('schluss', 'max'),
    min_kurs=('schluss', 'min'),
).round(2).reset_index()

summary['gesamtrendite_pct'] = (
    (summary['end_kurs'] - summary['start_kurs']) /
     summary['start_kurs'] * 100
).round(2)

os.makedirs("exports", exist_ok=True)
# In Excel speichern (mehrere Blatter)
with pd.ExcelWriter('exports/aktien_dashboard.xlsx', engine='openpyxl') as writer:
    summary.to_excel(writer, sheet_name='Uebersicht',  index=False)
    df.to_excel(writer,      sheet_name='Rohdaten',    index=False)
    df_kz.to_excel(writer,   sheet_name='Kennzahlen',  index=False)

print('Excel-Datei erstellt: exports/aktien_dashboard.xlsx')
```
